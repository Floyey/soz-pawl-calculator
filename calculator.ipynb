```python
# Prix de vente
PRICES = [160, 500, 1600, 6000]
LABELS = ["Banal", "Joli", "Sublime", "Divin"]

# Objectif de vente
SALES_TARGET = 300_000

# PrioritÃ© de vente (0=BANAL ; 1=JOLI ; 2=SUBLIME ; 3=DIVIN --- Le premier Ã©lÃ©ment du tableau est celui Ã  vendre en prioritÃ©)
SALES_PRIORITY = [0, 1, 2, 3]

# Nombre de combinaisons Ã  afficher
MAX_TOP_RESULTS = 5
```

```python
class Colors:
    RESET = "\033[0m"
    BOLD = "\033[1m"
    BLUE = "\033[94m"
    GREEN = "\033[92m"
    GRAY = "\033[90m"


def calculate_sales_combinations(stocks):
    # stocks = [stock_basic, stock_nice, stock_sublime, stock_divine]
    combinations = []

    price_basic, price_nice, price_sublime, price_divine = PRICES
    stock_basic, stock_nice, stock_sublime, stock_divine = stocks

    for nb_divine in range(min(stock_divine, SALES_TARGET // price_divine) + 1):
        remaining_after_divine = SALES_TARGET - nb_divine * price_divine

        for nb_sublime in range(min(stock_sublime, remaining_after_divine // price_sublime) + 1):
            remaining_after_sublime = (remaining_after_divine - nb_sublime * price_sublime)

            for nb_nice in range(min(stock_nice, remaining_after_sublime // price_nice) + 1):
                remaining_after_nice = remaining_after_sublime - nb_nice * price_nice

                if remaining_after_nice % price_basic == 0:
                    nb_basic = remaining_after_nice // price_basic
                    if nb_basic <= stock_basic:
                        combinations.append((nb_basic, nb_nice, nb_sublime, nb_divine))

    combinations.sort(key=lambda comb: tuple(-comb[i] for i in SALES_PRIORITY))
    return combinations


def display_combination(combination, number, is_top=False):
    basic, nice, sublime, divine = combination
    total = sum(q * p for q, p in zip(combination, PRICES))

    title_color = Colors.GREEN if is_top else Colors.BLUE
    prefix = "ğŸ† " if is_top else "ğŸ“‹ "
    width = 60

    print(f"{title_color}{Colors.BOLD}{prefix}Combinaison {number}:{Colors.RESET}")
    print("â”Œ" + "â”€" * (width - 2))

    line_fmt = "â”‚ {:<9} {:>3d} unitÃ©s x {:4d}$ = {:>10,d}$"
    print(line_fmt.format("Banal:", basic, PRICES[0], basic * PRICES[0]))
    print(line_fmt.format("Joli:", nice, PRICES[1], nice * PRICES[1]))
    print(line_fmt.format("Sublime:", sublime, PRICES[2], sublime * PRICES[2]))
    print(line_fmt.format("Divin:", divine, PRICES[3], divine * PRICES[3]))

    print("â”œ" + "â”€" * (width - 2))
    total_str = f"TOTAL:    {total:,d}$"
    print(f"â”‚ {Colors.BOLD}{total_str}{Colors.RESET}")

    print("â””" + "â”€" * (width - 2))
    print()


def display_combinations(combinations, limit=None):
    if not combinations:
        print(f"{Colors.GRAY}âŒ Aucune combinaison trouvÃ©e pour atteindre exactement {SALES_TARGET:,d}${Colors.RESET}")
        return

    nb_to_display = min(limit or len(combinations), len(combinations))
    title = f"ğŸ¯ TOP {nb_to_display} meilleures combinaisons" if limit else "ğŸ“Š Toutes les combinaisons"

    print(f"\n{Colors.BLUE}{Colors.BOLD}{'='*60}{Colors.RESET}")
    print(f"{Colors.BLUE}{Colors.BOLD}{title.center(60)}{Colors.RESET}")
    prio_txt = " > ".join(LABELS[i] for i in SALES_PRIORITY)
    print(f"{Colors.GRAY}Objectif: {SALES_TARGET:,d}$ | PrioritÃ©: {prio_txt}{Colors.RESET}")
    print(f"{Colors.BLUE}{Colors.BOLD}{'='*60}{Colors.RESET}\n")

    for i in range(nb_to_display):
        display_combination(combinations[i], i + 1, is_top=(i == 0 and limit is not None))


def read_int(prompt, min_value=0):
    while True:
        raw = input(prompt).strip()
        try:
            value = int(raw)
            if value < min_value:
                print(f"{Colors.GRAY}âš ï¸ Entrez un nombre >= {min_value}.{Colors.RESET}")
                continue
            return value
        except ValueError:
            print(f"{Colors.GRAY}âš ï¸ Entrez un nombre entier valide.{Colors.RESET}")


def main():
    print(f"{Colors.BLUE}{Colors.BOLD}")
    print("â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—")
    print("â•‘           ğŸª CALCULATEUR DE VENTE DE MEUBLES ğŸª         â•‘")
    print("â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•")
    print(f"{Colors.RESET}")

    print(f"ğŸ¯ Objectif de vente: {Colors.BOLD}{SALES_TARGET:,d}${Colors.RESET}")
    print(f"{Colors.GRAY}ğŸ“Š PrioritÃ© de vente: {' â†’ '.join(LABELS[i] for i in SALES_PRIORITY)}{Colors.RESET}\n")

    print(f"{Colors.GREEN}{Colors.BOLD}ğŸ“¦ SAISIE DES STOCKS DISPONIBLES{Colors.RESET}")
    print("â”€" * 40)

    stocks = [
        read_int(f"Stock meubles banals ({PRICES[0]}$ chacun): "),
        read_int(f"Stock meubles jolis ({PRICES[1]}$ chacun): "),
        read_int(f"Stock meubles sublimes ({PRICES[2]}$ chacun): "),
        read_int(f"Stock meubles divins ({PRICES[3]}$ chacun): "),
    ]

    print(f"\n{Colors.GRAY}ğŸ”„ Calcul des combinaisons en cours...{Colors.RESET}")
    combinations = calculate_sales_combinations(stocks)

    if not combinations:
        print(f"\n{Colors.GRAY}âŒ Aucune combinaison trouvÃ©e pour atteindre exactement {SALES_TARGET:,d}${Colors.RESET}")
        return

    display_combinations(combinations, limit=MAX_TOP_RESULTS)

    best = combinations[0]
    print(f"\n{Colors.GREEN}{Colors.BOLD}ğŸ“ˆ RÃ‰SUMÃ‰{Colors.RESET}")
    print("â”€" * 20)
    print(f"ğŸ¯ Objectif: {SALES_TARGET:,d}$")
    print(f"âœ… Combinaisons trouvÃ©es: {len(combinations)}")
    print("ğŸ† Meilleure combinaison (selon prioritÃ©):")
    print(f"   {best[0]} Banals + {best[1]} Jolis + {best[2]} Sublimes + {best[3]} Divins")


# Pas touche
if __name__ == "__main__":
    main()
```
